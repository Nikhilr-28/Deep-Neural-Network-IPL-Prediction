```python
import pandas as pd
from uuid import uuid4
```

### GATHERING ALL players

```python
players = pd.read_csv('./dataset/IPL_TEAMS - PLAYER_2025.csv')
```

```python
players.head(5)
```

```python
players['PLAYER_TYPE'].unique()
```

```python
players.columns
```

```python
players_id = []
for i in range(players['PLAYER_NAME'].count()):
    players_id.append(str(uuid4()))
```

```python
players['PLAYER_ID'] = players_id
```

```python
match_cols = [
    'IPL', 'T20I', 'ODI', 'STATE_T20', 'SMAT_ELITE',
    'VIJAY_HAZARE', 'VINOO_MANKAD', 'DULEEP_TROPHY',
    'RANJI_TROPHY', 'BBL', 'CPL', 'SA20', 'ILT20'
]
MASTER_DATA = []

for player in players.iterrows():
    
    temp = []
    for i in match_cols:
        data = {
            'PLAYER_ID': player[1][3]+'_'+i,
            'PLAYER_NAME': player[1][0],
            'PLAYER_TYPE': player[1][1],
            'PLAYER_IPL_TEAM': player[1][2],
        }
        temp.append(data)

    MASTER_DATA.extend(temp)
```

```python
players = pd.DataFrame(MASTER_DATA)
```

```python
players
```

```python
players.set_index('PLAYER_ID', inplace=True)
```
