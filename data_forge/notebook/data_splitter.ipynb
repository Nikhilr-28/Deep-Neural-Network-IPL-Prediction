#### Splitting Dataset template between Nikhil, atharva and Gaurav

(i) Nikhil Dataset

```python
NIKHIL_BATSMAN_IPL = batsman[(batsman["PLAYER_IPL_TEAM"] == 'CSK' ) | (batsman["PLAYER_IPL_TEAM"] == 'GT' ) | (batsman['PLAYER_IPL_TEAM'] == 'PK') | (batsman['PLAYER_IPL_TEAM'] == 'RR')]
```

```python
NIKHIL_BATSMAN_IPL.to_csv('../dataset/nikhil/NIKHIL_IPL_BATSMAN.csv')
```

```python
NIKHIL_BOWLERS_IPL = bowlers[(bowlers["PLAYER_IPL_TEAM"] == 'CSK' ) | (bowlers["PLAYER_IPL_TEAM"] == 'GT' ) | (bowlers['PLAYER_IPL_TEAM'] == 'PK') | (bowlers['PLAYER_IPL_TEAM'] == 'RR')]
```

```python
NIKHIL_BOWLERS_IPL.to_csv('../dataset/nikhil/NIKHIL_BOWLERS_IPL.csv')
```

(ii) Atharva Dataset

```python
ATTA_BATSMAN_IPL = batsman[(batsman["PLAYER_IPL_TEAM"] == 'MI' ) | (batsman["PLAYER_IPL_TEAM"] == 'RCB' ) | (batsman['PLAYER_IPL_TEAM'] == 'KKR')]
```

```python
ATTA_BATSMAN_IPL.to_csv('../dataset/atta/ATTA_BATSMAN_IPL.csv')
```

```python
ATTA_BOWLERS_IPL = bowlers[(bowlers["PLAYER_IPL_TEAM"] == 'MI' ) | (bowlers["PLAYER_IPL_TEAM"] == 'RCB' ) | (bowlers['PLAYER_IPL_TEAM'] == 'KKR')]
```

```python
ATTA_BOWLERS_IPL.to_csv('../dataset/atta/ATTA_BOWLERS_IPL.csv')
```

(iii) Gaurav Dataset

```python
GAURAV_BATSMAN_IPL = batsman[(batsman["PLAYER_IPL_TEAM"] == 'SRH' ) | (batsman["PLAYER_IPL_TEAM"] == 'LSG' ) | (batsman['PLAYER_IPL_TEAM'] == 'DC')]
```

```python
GAURAV_BATSMAN_IPL.to_csv('../dataset/gaurav/GAURAV_BATSMAN_IPL.csv')
```

```python
GAURAV_BOWLERS_IPL = bowlers[(bowlers["PLAYER_IPL_TEAM"] == 'SRH' ) | (bowlers["PLAYER_IPL_TEAM"] == 'LSG' ) | (bowlers['PLAYER_IPL_TEAM'] == 'DC')]
```

```python
GAURAV_BOWLERS_IPL.to_csv('../dataset/gaurav/GAURAV_BOWLERS_IPL.csv')
```
